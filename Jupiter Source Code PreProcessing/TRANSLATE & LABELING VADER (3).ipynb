```python
import pandas as pd
import googletrans
from googletrans import Translator
```

```python
df = pd.read_csv('DATA_PREPROCESSING_INDIHOME_PANDEMI (3).csv')
df.head(10)
```

```python
translator = Translator()
translations = {}
for column in df.columns:
    #unique elements of the column
    unique_elements = df[column].unique()
    for element in unique_elements:
        #adding all the translations to a dictionary (translations)
        translations[element] = translator.translate(element).text
translations
```

```python
# Mengganti semua kata yang sudah ditranslate
df.replace(translations, inplace = True)
df.head(10)
```

```python
df.to_csv('DATA_TRANSLATE_PANDEMI.csv', encoding='utf-8', index=False)
```

```python
import pandas as pd
def load_data():
    data = pd.read_csv('DATA_TRANSLATE_PANDEMI.csv')
    return data
```

```python
tweet_df = load_data()
tweet_df.head()
```

```python
!pip install VaderSentiment
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
scores = [analyser.polarity_scores(x) for x in tweet_df['Tweet_bersih']]
print(scores)
tweet_df['Compound_Score'] = [x['compound'] for x in scores]
```

```python
tweet_df.head()
```

```python
tweet_df.nsmallest(10,['Compound_Score'])
```

```python
#Compound Score Lexicon Based
tweet_df.loc[tweet_df['Compound_Score'] < 0, 'Sentiments'] = '0' #Negatif
tweet_df.loc[tweet_df['Compound_Score'] == 0, 'Sentiments'] = '2'#Netral
tweet_df.loc[tweet_df['Compound_Score'] > 0, 'Sentiments'] = '1' #Positif

tweet_df
```

```python
tweet_df.drop(index=tweet_df[tweet_df['Sentiments']=='2'].index, inplace=True)
tweet_df
```

```python
tweet_df.to_csv('HASIL_LABELING_PANDEMI.csv', encoding='utf-8', index=False)
```
