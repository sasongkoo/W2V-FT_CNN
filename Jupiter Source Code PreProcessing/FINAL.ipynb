```python
def load_data():
    data = pd.read_csv('DATA_PREPROCESSING_INDIHOME_PANDEMI_TRANSLATE.csv')
    return data
df = load_data()
df
```

```python
#Tahap Cleaning
def cleaning(tweet):
    import re
    import string

def cleaning(tweet):
    if not isinstance(tweet, (str, bytes)):
        tweet = str(tweet)
    # Lowercase
    tweet = tweet.lower()
    # Remove numbers
    tweet = re.sub('\d+', ' ', tweet)
    # Remove punctuation
    tweet = tweet.translate(str.maketrans('', '', string.punctuation))
    #Menghapus karakter spesial
    tweet=re.sub(r'[^\x00-\x7F]+',' ',tweet)
    # Menghapus link
    tweet = re.sub(r'https:\/\/.*',' ', tweet)
    # Menghapus mention
    tweet = re.sub(r'@[\w\d]*', ' ', tweet)
    # Menghapus hashtag
    tweet = re.sub(r'#(\S+)', ' ', tweet)
    #Menghapus RT
    tweet = re.sub(r'^rt[\S]+', ' ', tweet)
    #Menghapus Emoji
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
    tweet=emoji_pattern.sub(r'',tweet)
    #Menghilangkan white space
    tweet= ' '.join(tweet.split())
    return tweet

df['cleaning']=df['translated_indonesia'].apply(lambda x: cleaning(x))
df
```

```python
#Tahap Tokenizing
def tokenizing(tweet):
    tweet = nltk.tokenize.word_tokenize(tweet)
    return tweet
df['tokenizing'] = df['cleaning'].apply(lambda x: tokenizing(x))
df
```

```python
normalized_word = pd.read_csv("kamus_normalisasi.csv", encoding='latin1')

normalized_word_dict={}
for index, row in normalized_word.iterrows():
    if row[0] not in normalized_word_dict:
        normalized_word_dict[row[0]] = row[1]

def normalized_term(document):
    return [normalized_word_dict[term] if term in normalized_word_dict else term for term in document]

df['normalisasi'] = df['tokenizing'].apply(lambda x: normalized_term(x))
df
```

```python
df.to_csv('DATA_PREPROCESSING_INDIHOME_PANDEMI_Normalisasi.csv',encoding='utf8', index=False)
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}

tweet = df['normalisasi']
for doc in tweet:
    for term in doc:
        if term not in term_dict:
            term_dict[term] = ' '
    print(len(term_dict))
    print("----------------------------------")
    
for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    print(term,":", term_dict[term])
    
    print(len(term_dict))
    print("----------------------------------")
```

```python
def get_stemmed_term(doc):
    return [term_dict[term] for term in doc]

df['stemming'] = df['normalisasi'].apply(lambda x: get_stemmed_term(x))
df
```

```python
def remove_punct(tweet):
    tweet  = " ".join([char for char in tweet if char not in string.punctuation])
    return tweet
df['Tweet_bersih'] = df['stemming'].apply(lambda x: remove_punct(x))
df
```
