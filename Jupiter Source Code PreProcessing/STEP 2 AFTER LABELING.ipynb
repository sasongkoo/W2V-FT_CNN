```python
import pandas as pd
def load_data():
    data = pd.read_csv('HASIL_LABELING_PANDEMI(UPDATE).csv')
    return data
```

```python
tweet_df = load_data()
tweet_df.head()
```

```python
tweet_df.drop(index=tweet_df[tweet_df['Compound_Score']=='0.0000'].index, inplace=True)
tweet_df
```

```python
tweet_df.loc[tweet_df['Compound_Score'] < 0, 'Label'] = '0' #Negatif
tweet_df.loc[tweet_df['Compound_Score'] == 0, 'Label'] = '2'#Netral
tweet_df.loc[tweet_df['Compound_Score'] > 0, 'Label'] = '1' #Positif

tweet_df
```

```python
tweet_df.drop(index=tweet_df[tweet_df['Label']=='2'].index, inplace=True)
tweet_df
```

```python
tweet_df.drop(tweet_df.columns[[2]], axis = 1, inplace = True)
tweet_df
```

```python
tweet_df.to_csv('HASIL_LABELING_PANDEMI(COMPOUND).csv', encoding='utf-8', index=False)
```
